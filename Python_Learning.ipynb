**Variables**

Variables store data in a computer memory. A variable refers to a memory address in which data is stored.

Python has several rules for naming variables:
1. must start with a letter or the underscore
2. cannot start with a number
3. no special characters except '_' can be used
4. case sensitive




```python
name = input("Provide your name") # taking input from user and declaring it to name variable
age = input("Your age") # input and variable declaration

print(name)
print(age)
```
